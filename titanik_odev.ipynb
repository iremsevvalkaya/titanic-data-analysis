```python
import pandas as pd
import numpy as np
```

```python
df_titanic = pd.read_csv('Titanic_dataset.csv')
df_titanic.head()
```

```python
#1.soru Kazada ölenlerin yaş ortalamasını bulunuz
df_dead = df_titanic[df_titanic['Survived'] == 0]
dead_age_mean = df_dead['Age'].mean()
float(dead_age_mean)
```

```python
#2.soru Kazada ölenlerin bilet fiyatlarının ortalamasını ve medyanını bulunuz
dead_fare_mean = df_dead['Fare'].mean()
dead_fare_median = df_dead['Fare'].median()
```

```python
float(dead_fare_mean)
```

```python
float(dead_fare_median)
```

```python
#3.soru Kazada ölen erkeklerin yaş ortalamasını bulunuz
dead_men = df_dead[df_dead['Sex'] == 'male']
dead_men_age_mean = dead_men['Age'].mean()
float(dead_men_age_mean)
```

```python
#4.soru Kazada ölen Kadınların yaş ortalamasını bulunuz
dead_women = df_dead[df_dead['Sex'] == 'female']
dead_women_age_mean = dead_women['Age'].mean()
float(dead_women_age_mean)
```

```python
#5.soru Kazadan kurtulanların yaş ortalamasını bulunuz
df_survived = df_titanic[df_titanic['Survived'] == 1]
survived_age_mean = df_survived['Age'].mean()
float(survived_age_mean)
```

```python
#6.soru Kazadan kurtulanların bilet fiyatlarının ortalamasını bulunuz
survived_fare_mean = df_survived['Fare'].mean()
float(survived_fare_mean)
```

```python
#7.soru Kazadan kurtulan toplam kişi sayısını bulunuz
survivors_count = len(df_survived)
survivors_count
```

```python
#8.soru 10 yaşından küçüklerin bilet fiyatlarının medyan değerini bulunuz
children_under_10 = df_titanic[df_titanic['Age'] < 10]
children_fare_median = children_under_10['Fare'].median()
children_fare_median
```

```python
#9.soru 1.sınıf, 2.sınıf ve 3.sınıf bilet fiyatlarının ortalama ve medyanlarını karşılaştırınız.
fare_stats_by_class = df_titanic.groupby('Pclass')['Fare'].agg(['mean', 'median']).rename_axis('Yolcu Sınıfı')
fare_stats_by_class
```

```python
#Birinci sınıf yolcular çok daha yüksek ücret ödemiştir.
#Üçüncü sınıf, en ekonomik sınıf olarak ortalama ve medyan açısından oldukça düşüktür.
#Bu farklar, sosyal sınıflar arasındaki ekonomik uçurumu da yansıtır.
```
